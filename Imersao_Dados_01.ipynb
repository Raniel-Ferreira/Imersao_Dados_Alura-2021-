#**Desafio 01 :Investigar por que a classe tratamento é tão desbalanceada?**


Na tabela existem duas classificações na coluna tratamento

````python
dados['tratamento'].value_counts()
````

````python
com_droga = dados[dados['tratamento'] == 'com_droga']
com_droga['droga'].unique()

````

````python
com_controle = dados[dados['tratamento'] == 'com_controle']
com_controle['droga'].unique()
````

*Resposta*:\
Na primeira célula, vemos a contagem de quem foi tratado com o remédio e quem foi parte do grupo controle e já dá pra notar uma diferença enorme nos valores.
\
Seguindo adiante coloquei códigos pra saber quais eram os tipos de drogas que estavam em cada linha (as letras estão mascaradas de propósito) e dá pra ver que quem recebeu o remédio tomou varias doses diferentes, enquanto que quem estava no controle recebeu somente um tipo de dose.
\
Isso se dá porque quando se fazem pesquisas com pessoas, no âmbito da saúde em geral, precisamos ter um grupo que vai sofrer uma intervenção e outro grupo(denominado grupo controle) que vai receber uma intervenção "mascarada"(um placebo).
\
No caso das drogas, o grupo que participou da pesquisa que não era o controle recebeu diversas doses diferentes pra que se fosse avaliado como eles iriam reagir. Já o outro grupo recebeu uma droga "placebo" (é como se estivessem tomando água saborizada achando que é um remédio)
\
\
Isso é muito importante quando se realizam pesquisas na área da saúde.

#**Desafio 03: Proporção das classes de tratamento**

Pra descobrir a proporção podemos utilizar o parâmetro *normalize* dentro da função *value_counts* (aquela que usamos pra retornar quantos valores tinha no desafio 01)

````python
dados['tratamento'].value_counts(normalize = True)
````

Pra deixar mais legível vamos criar outra variável e multiplicar por 100

````python
percentual = dados['tratamento'].value_counts(normalize = True) * 100 
percentual
````

Já está muito bom mas dá pra deixar mais ainda, podemos utilizar a função *round()* que arredonda os números e dentro do parênteses podemos colocar quantas casas depois da vírgula queremos que apareça:
 

````python
percentual = dados['tratamento'].value_counts(normalize = True) * 100 
percentual.round(2)
````

*Reposta*: 92,16% das drogas estavam no grupo de experimento e 7,84% no grupo controle.

#**Desafio 05: Procurar na documentação do Pandas o método query**

*Resposta*:
Query vem português significa: inquerir, perguntar, tirar um dúvida. 
\
\
O método *query()* da biblioteca pandas serve para nós pedirmos ao nosso computador que ele mostre algo dentro dos dados que nós coletamos. Dentro dos parênteses escrevemos o que gostaríamos que ele retornasse. 
\
\
Vejamos o exemplo:

````python
dados.query("tempo > 24")
#aqui ele só retornou as linhas onde o tempo foi maior que 24
````

````python
dados.query('tempo > 24 and tempo < 72')
#aqui eu pedi todas as linhas onde o tempo era maior que 24h e menor que 72
````

````python
dados.query('tempo > 72')
#aqui não retornamos nada porque não teve nenhum tratamento que passou de 72h
````

````python
dados.query('`g-0` > 0')
#temos que tomar cuidado com misturas de letras e números nesse caso g-0 tem os dois por isso precisamos colocar as crases ` ` pra o computador que aquele g-0 é um texto 
````

#**Desafio 06: Renomear as colunas tirando o hífen.**

*Resposta*:
Pra renomear as colunas existe uma função simples chamada *rename()*. Onde colocamos o nome do que queremos renomear, dois pontos e o novo nome; também precisamos especificar se é uma linha (index) ou coluna(columns).

```rename(columns={'g-0' : 'g0'})```
\
\
Também existe o *rename_axis()*, nesse caso ele dá um nome novo para um eixo(axis) da tabela.

````python
dados.rename(columns={'g-0' : 'g0'})
#perceba que mudou apenas a coluna do g-0
````

````python
dados.rename_axis("Índices")
#aqui ele acrescentou mais uma linha/eixo na tabela
````
